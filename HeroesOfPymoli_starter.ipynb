* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
 

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# total number of players 
total = purchase_data["SN"].nunique(dropna=True)

total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python

# number of unique items
unique_items = purchase_data["Item ID"].nunique(dropna=True)
# Average Price
average_price = purchase_data["Price"].mean()
# number of purchases
purchase_count = purchase_data["Purchase ID"].count()
#total revenue
revenue_total = purchase_data["Price"].sum()
# Creating Dataframe
Analysis = pd.DataFrame([{
    "Number of Unique Items": unique_items, "Average Price": average_price, "Number of Purchases": purchase_count, "Total Revenue": revenue_total
    }])
# Changing the format
Analysis["Average Price"]=Analysis["Average Price"].astype(float).map("${:,.2f}".format)
Analysis["Total Revenue"]= Analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
#calling function
Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# filter out the duplicates
unique = purchase_data.drop_duplicates(["SN"])
# Count the number of players in each gender group
gender_count = unique["Gender"].value_counts()
# calculating percentages of gender
Percentages = (unique["Gender"].value_counts()/unique["Gender"].count())*100

# Creating a Dataframe
gender_Analysis = pd.DataFrame({"Number of players": gender_count, "Percentage": Percentages})
# Formating for percentage
gender_Analysis["Percentage"]=gender_Analysis["Percentage"].astype(float).map("{:,.2f}%".format)
gender_Analysis.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group by gender
grouped_by_gender = purchase_data.groupby(['Gender'])
# average price
Purchase_mean = grouped_by_gender["Price"].mean()
#Purchase count
Purchase_count = grouped_by_gender["Purchase ID"].count()
# Purchase total
Purchase_total = grouped_by_gender["Price"].sum()
Player_total = grouped_by_gender["SN"].count()
PPP = Purchase_total/gender_count
purchase_Analysis = pd.DataFrame({"Purchase Count": Purchase_count, "Average Price": Purchase_mean, "Total Purchases": Purchase_total, "Avg Total Purchase per Person": PPP})
purchase_Analysis["Total Purchases"]= purchase_Analysis["Total Purchases"].astype(float).map("${:,.2f}".format)
purchase_Analysis["Average Price"]= purchase_Analysis["Average Price"].astype(float).map("${:,.2f}".format)
purchase_Analysis["Avg Total Purchase per Person"]= purchase_Analysis["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
purchase_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins for age range
bins = [0, 9, 14, 19, 24, 29,
        34, 39,100]

# Create labels for these bins
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]


age_range = pd.cut(unique["Age"], bins, labels=age_labels).value_counts()
age_range


#age_count = unique["Age"].value_counts()
#player_age = purchase_data["Age"]

# calculating percentages of each age range 
Percentages_age = (age_range/total)*100

# Creating a Dataframe
age_Analysis = pd.DataFrame({"Number of players": age_range, "Percentage": Percentages_age})
# Formating for percentage
age_Analysis["Percentage"]=age_Analysis["Percentage"].astype(float).map("{:,.2f}%".format)
age_Analysis
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# creating new column for groupby
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=age_labels)
# Groupby age
grouped_by_age = purchase_data.groupby(["Age Range"])

Purchase_mean_age = grouped_by_age["Price"].mean()
#Purchase count
Purchase_count_age = grouped_by_age["Purchase ID"].count()
# Purchase total
Purchase_total_age = grouped_by_age["Price"].sum()
Player_total_age = grouped_by_age["SN"].count()
PPP_age = Purchase_total_age/Player_total_age
purchase_Analysis_age = pd.DataFrame({"Purchase Count": Purchase_count_age, "Average Price": Purchase_mean_age, "Total Purchases": Purchase_total_age, "Avg Total Purchase per Person": PPP_age})
purchase_Analysis_age["Total Purchases"]= purchase_Analysis_age["Total Purchases"].astype(float).map("${:,.2f}".format)
purchase_Analysis_age["Average Price"]= purchase_Analysis_age["Average Price"].astype(float).map("${:,.2f}".format)
purchase_Analysis_age["Avg Total Purchase per Person"]= purchase_Analysis_age["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
purchase_Analysis_age
```
